```python
# Execution  /modifier le type d'execution / GPU
import flair
import csv
from tqdm import tqdm
flair_sentiment = flair.models.TextClassifier.load('en-sentiment')


predict_result = []
with open("IMDB-Dataset.100.csv", newline='') as csvfile:
  reader = csv.reader(csvfile, delimiter=',', quotechar='"') 
  # skip column names
  next(reader)
  polarities=dict()
  occurences=dict()
  row_num=0
  for row in reader:
    row_num+=1
    # Pour fixer à 20 le nombre de commentaires à analyser
    if row_num >20:
      break
    s = flair.data.Sentence(row[0])
    flair_sentiment.predict(s)
    sentence_score=s.labels[0].score
    sent=[tok.text for tok in s]
    print(f"Phrase {row_num} : {row[0]} , {s.labels[0].value},  score : {sentence_score}")
    predict_result.append([row_num, s.labels[0].value, row[1]])
    for pos in tqdm(range(0, len(s.tokens))):
      word=sent[pos]
      substracted=flair.data.Sentence(" ".join(word for word in sent[:pos]+sent[pos+1:]))
      flair_sentiment.predict(substracted)
      substracted_score=substracted.labels[0].score
      delta=sentence_score-substracted_score
      print(f"Mot retiré : {word}")
      print(f"Polarité après retrait : {substracted.labels[0].value} (Confidence: {substracted_score})")
      print(f"Différence de polarité : {delta}\n")
      if word not in polarities:
        polarities[word]=delta
        occurences[word]=1
      else:
        polarities[word]+=delta
        occurences[word]+=1
    for word in polarities.keys():
        # est ce que comparer la myenne est une bonne idée ? car les mots qui apparaissent plus souvent auront plus de poids
      polarities[word]/=occurences[word]
    sorted_polarities=dict(sorted(polarities.items(), key=lambda item: item[1]))
  print(predict_result)
with open("result.txt", "w") as f:
  for word, polarity in sorted_polarities.items():
    print(word, round(polarity,3), occurences[word])
    f.write(word + " " + str(round(polarity,3)) + " " + str(occurences[word]) + "\n")
```
